```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('forestfire.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='day', palette='Set2', order=df['day'].value_counts().index)
plt.title('Number of Fires by Day')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='temp', bins=10, edgecolor='black')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='month', y='wind', palette='coolwarm', order=df['month'].value_counts().index)
plt.title('Wind Speed by Month')
plt.xlabel('Month')
plt.ylabel('Wind Speed')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='temp', y='area', hue='month', palette='Spectral')
plt.title('Area Burned vs Temperature')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Area Burned (ha)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='month', palette='pastel', order=df['month'].value_counts().index)
plt.title('Fire Counts by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,8))
corr = df[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
